```python
# Установка библиотеки (раскомментируйте и выполните один раз)
# !pip install -U yandex-music

from yandex_music import Client
import os

# -----------------------------
# Настройки
# -----------------------------

TOKEN = "ваш_токен_от_Яндекс_Музыка"  # замените на свой токен
PLAYLIST_URL = "https://music.yandex.ru/playlists/b492efec-af96-6e43-b003-32d88d1abb31?ref=mysanta.ru"
DOWNLOAD_DIR = "yandex_playlist"  # папка для сохранения треков

# Создаем папку, если не существует
os.makedirs(DOWNLOAD_DIR, exist_ok=True)

# -----------------------------
# Инициализация клиента
# -----------------------------
client = Client(TOKEN).init()

# Получаем ID пользователя и плейлиста из ссылки
# Ссылка имеет формат: https://music.yandex.ru/users/<user_id>/playlists/<playlist_id>
# Иногда ссылка может быть сокращенной, тогда библиотека сама определяет плейлист
playlist = client.playlist_from_url(PLAYLIST_URL)

# -----------------------------
# Скачивание треков
# -----------------------------
print(f"Плейлист: {playlist.title}")
print(f"Треков: {len(playlist.tracks)}")

for track_info in playlist.tracks:
    track = track_info.track.fetch_track()  # получаем полный объект трека
    filename = f"{track.artists[0].name} - {track.title}.mp3"
    filepath = os.path.join(DOWNLOAD_DIR, filename)
    print(f"Скачиваем: {filename}")
    track.download(filepath)

print("Все треки скачаны!")
```
